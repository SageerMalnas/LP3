```python
# Step 3: Normalize the train and test data (standardization)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**Encoding categorical (string based) data.**

```python
# Step 4: Initialize and build the neural network model
model = Sequential()
```

```python
# You can experiment with the number of layers and neurons based on your dataset.
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

**Split the countries into respective dimensions. Converting the string features into their own dimensions.**

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

**Dimensionality reduction. A 0 on two countries means that the country has to be the one variable which wasn't included**

```python
# Fit the model to the training data
model.fit(X_train, y_train, epochs=20, batch_size=32)
```

# **Splitting the Dataset**
Training and Test Set

```python
# Step 5: Evaluate the model
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)  # Convert probabilities to binary predictions

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)

# Calculate confusion matrix
confusion = confusion_matrix(y_test, y_pred)

# Print the results
print("Accuracy:", accuracy)
print("Confusion Matrix:")
print(confusion)
```
