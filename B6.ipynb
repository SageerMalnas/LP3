```python
# Select the relevant features for clustering (e.g., numerical columns)
numeric_data = df.select_dtypes(include=[np.number]).dropna()
```

```python
# Standardize the features (important for K-Means)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(numeric_data)
```

```python
# Use the Elbow method to determine the number of clusters
inertia = []  # Sum of squared distances to the nearest cluster center
```

```python
for i in range(1, 11):  # Trying different values of k (number of clusters)
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# for i in range(1,11):
#   kmeans = KMeans(n_clusters=i,random_state=0)
#   kmeans.fit(scaled_features)
#   inertia.append(kmeans.inertia_)
```

```python
# Plot the Elbow method graph
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```
